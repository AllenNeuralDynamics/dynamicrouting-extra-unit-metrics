```python
# import numpy as np
# import numpy.typing as npt 
import tqdm
import polars as pl

import concurrent.futures
import multiprocessing

with concurrent.futures.ProcessPoolExecutor(max_workers=2, mp_context=multiprocessing.get_context('spawn')) as executor:
    results = []
    for session_id in utils.get_df('units')['session_id'].unique():
        futures = []
        futures.append(executor.submit(utils.get_spike_counts_by_trial, session_id))
    for future in tqdm.tqdm(concurrent.futures.as_completed(futures), total=len(futures), unit='sessions'):
        results.append(future.result())

result = pl.concat(results)
```
